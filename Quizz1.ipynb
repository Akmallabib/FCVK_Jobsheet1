3. Membuat garis vertikal, horizontal

```python
# inisialisasi gambar
gambar1 = img

# ambil gambar
urls = [gambar1]
for url in urls:
  image = img
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)  #convert color to RGB
  final_frame = cv.hconcat((image, image_2))      #concatenate image

# menampilkan tinggi dan lebar gambar
  tinggi = image_2.shape[0]
  lebar = image_2.shape[1]
  print("resolusi image: tinggi x lebar = ",tinggi," x ",lebar)
  cv2_imshow(final_frame)                         #show image
  print('\n')
```

```python
# mengambil gambar
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis horizontal
for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

# menampilkan gambar
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
# mengambil gambar
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis vertikal
for x in range (tinggi):
  image_3[x, int((lebar)/2)] = [255,255,255]

# menampilkan gambar
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

4. Flip vertikal / horizontal

```python
# Vertikal
img_flip = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB), 0)
plt.imshow(img_flip)
```

```python
# Horizontal
img_flip = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB), 1)
plt.imshow(img_flip)
```

5. Menambah shape

```python
# inisialisasi gambar
img2 = img
img3 = cv.cvtColor(img,cv.COLOR_BGR2RGB)

# proses penambahan shape

# shape rectagle
cv.rectangle(img3,pt1=(150,150),pt2=(250,80),color=(255,0,0),thickness=5) # red
cv.rectangle(img3,pt1=(390,150),pt2=(480,80),color=(0,255,0),thickness=5) # green
cv.rectangle(img3,pt1=(630,150),pt2=(720,80),color=(0,0,255),thickness=5) # blue

# shape line
cv.line(img3,pt1=(250,100),pt2=(300,80),color=(0,255,255),thickness=10) # tosca
cv.line(img3,pt1=(100,80),pt2=(150,100),color=(0,255,255),thickness=10) # tosca
cv.line(img3,pt1=(450,180),pt2=(450,150),color=(0,255,255),thickness=10) # tosca

# shape circle
cv.circle(img3,center=(570,50),radius=30,color=(0,0,0),thickness=8) # hitam
cv.circle(img3,center=(790,50),radius=30,color=(0,0,0),thickness=8) # hitam

# menampilkan gambar
plt.imshow(img3)
```
